```python
import random

# Maze configuration
maze = [
    list("🧱🧱🧱🧱🧱🧱🧱🧱🧱🧱"),
    list("🧱🚶     🧱  🚪🧱"),
    list("🧱 🧱🧱 🧱 🧱🧱🧱🧱"),
    list("🧱    🧱     🧱"),
    list("🧱 🧱 🧱 🧱 🧱 🧱"),
    list("🧱     🧱      🧱"),
    list("🧱🧱🧱🧱🧱🧱🧱🧱🧱🧱")
]

# Find player position
def find_player(maze):
    for i, row in enumerate(maze):
        for j, val in enumerate(row):
            if val == '🚶':
                return i, j

# Display maze
def print_maze():
    for row in maze:
        print("".join(row))

# Movement
def move_player(dir_key):
    global maze
    y, x = find_player(maze)
    moves = {'W': (-1, 0), 'S': (1, 0), 'A': (0, -1), 'D': (0, 1)}
    
    if dir_key not in moves:
        print("⚠️ Invalid input! Use W, A, S, D.")
        return True
    
    dy, dx = moves[dir_key]
    ny, nx = y + dy, x + dx
    
    if maze[ny][nx] == '🧱':
        print("🚫 Bumped into a wall!")
        return True
    elif maze[ny][nx] == '🚪':
        print("🎉 Congratulations! You escaped the maze!")
        return False
    
    # Move player
    maze[y][x] = ' '
    maze[ny][nx] = '🚶'
    return True

# Game loop
print("🎮 Welcome to Maze Runner!")
print("🚶 = You, 🧱 = Wall, 🚪 = Exit")
print("🧭 Controls: W = up, S = down, A = left, D = right\n")

alive = True
while alive:
    print_maze()
    move = input("Move (W/A/S/D): ").strip().upper()
    alive = move_player(move)
    print("\n" + "-" * 20)
```
